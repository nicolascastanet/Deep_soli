```python
from net import CNNModel
import torch
import numpy as np
import matplotlib.pyplot as plt
import pickle
import seaborn as sn
```

# Load model

```python
conf_matr = pickle.load( open("res/conf1.p", "rb"))
loss = pickle.load( open( "res/loss.p", "rb" ))
accuracy = pickle.load( open( "res/acc_list.p", "rb" ))
```

# Loss / Accuracy

```python
x = np.arange(len(loss))
y1 = loss
y2 = accuracy

fig = plt.figure()
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

ax1.set_ylabel("Loss")
ax2.set_ylabel("Acuuracy")

plt.subplot(211)
plt.plot(x, y1)
plt.subplot(212)
plt.plot(x,y2)

plt.show()
```

# Confusion matrix

```python
confusion_matrix = conf_matr/conf_matr.sum(1)
fig, ax = plt.subplots(figsize=(12,7)) 
ax = sn.heatmap(confusion_matrix, annot=True, linewidths = .8, cmap="YlGnBu")
plt.show()
```
